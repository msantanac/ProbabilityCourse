# Ejemplos de funciones de probabilidad

En este notebook se muestra como graficar funciones de probabilidad y funciones de distribución acumuladas de variables aleatorias discretas.

Suponga que se tiene una varible aleatoria $X$ discreta, con función de probabilidad $f(x)$, dada por la siguiente tabla:

| x   |   1 |   2 |   3 |   4 |   5 |
|-----|-----|-----|-----|-----|-----|
|f(x) | 0.1 | 0.2 | 0.4 | 0.2 | 0.1 |

Nos gustaría visualizar la gráfica de la función de probabilidad $f(x)$ y la gráfica de la función de distribución acumulada $F(x)$.
Comenzamos cargando las librerías correspondientes.

```python
import numpy as np
import matplotlib.pyplot as plt
```

Ahora cargamos los datos.

```python
# Valores de la variable
x = np.array([1,2,3,4,5])
# Valores de la función de probabilidad
fx = np.array([0.1,0.2,0.4,0.2,0.1])
```

Creamos la gráfica de la función de probabilidad primero.

```python
# Gráfica de la función de probabiliad
plt.figure()
plt.stem(x, fx, "-") # gráfica de líneas verticales
plt.xlabel(r"$x$")
plt.ylabel(r"$f(x)$")
plt.title("Función de probabilidad")
```

Ahora generamos la funcion de probabilidad acumulada $F(x)$.

```python
# CDF

# Creamos un vector auxiliar que tenga dos valores de la variable extra, uno
# antes del primer valor y otro después del último. Esto nos permitirá apresiar 
# mejor que la gráfica de la función de distribución es una función constante a
# pedazos, la cual es cero antes del primer valor de la variable, y uno después
# del último valor.

x_aux = np.concatenate((np.zeros(1), x, np.array([x[-1]])+1), out = None)
Fx_aux = np.concatenate((np.zeros(1), np.cumsum(fx), np.ones(1)), axis = 0)
plt.figure()
plt.step(x_aux, Fx_aux, where = "post")
plt.xlabel(r"$x$")
plt.ylabel(r"$F(x)$")
plt.title(R"Función de distribución acumulada $F(x)$")
```
